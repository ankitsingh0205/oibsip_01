```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
dframe_Iris= pd.read_csv('Iris.csv')
dframe_Iris .head()
```

```python
#Dropping Useless Columns
dframe_Iris=dframe_Iris.drop(columns=["Id"])
```

```python
dframe_Iris.head()
```

```python
dframe_Iris["Species"].replace({"Iris-setosa":1,"Iris-versicolor":2,"Iris-virginica":3},inplace=True)
```

```python
dframe_Iris
```

```python
#Creating Arrays
x=pd.DataFrame(dframe_Iris,columns=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]).values
```

```python
x
```

```python
y=dframe_Iris.Species.values.reshape(-1,1)
```

```python
y
```

```python
#Train Test Split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
k=6
knclr=KNeighborsClassifier(k)
```

```python
#Train The Model Using KNN
knclr.fit(x_train,y_train)
```

```python
y_pred=knclr.predict(x_test)
```

```python
y_pred=knclr.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_pred)*100
```
